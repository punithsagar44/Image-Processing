```python
import numpy as np

import cv2

import matplotlib.pyplot as plt

# Open the image.

im = cv2.imread( 'cat_damaged.png')
plt. imshow(im)

plt.show()

# Load the mask.

mask = cv2.imread('cat_mask.png', 0)
plt. imshow(mask)

plt.show()

# Inpaint.

dst = cv2.inpaint(im, mask, 2, cv2.INPAINT_TELEA)

# write the output.
cv2.imwrite( 'dimage_inpainted.png', dst)
plt.imshow(dst)

plt.show()
```
